## 파이썬에서도 정규 표현식 사용

```python
import re
```

```python
regex = r"([\w]+\s*:?\s*)?\(?\+?([0-9]{1,3})?\-?[0-9]{2,3}(\)|\-)?[0-9]{3,4}\-?[0-9]{4}"
```

```python
x = "Ki: +82-10-1234-5678"
```

```python
re.sub(regex, "REMOVED", x)
```

```python
x = "CONTENT jiu 02)1234-5678"
```

```python
re.sub(regex, "REMOVED", x)
```

## 치환자 사용

```python
x = """abcdefg
12345
ab12
a1bc2d
12ab
a1b
1ab
1a2
a1
1a
jjklmnop"""

regex = r"([a-z])[0-9]+([a-z])"
to = r"\1\2"

y = "\n".join([re.sub(regex, to, x_i) for x_i in x.split("\n")])
```

```python
print(y)
```
